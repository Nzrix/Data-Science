```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
plt.figure(figsize=[4,2])
sns.histplot(tips['tip'],bins = 30, kde = True)
```

```python
plt.figure(figsize=[4,2])
sns.countplot(data = tips, x = 'time')
```

```python
plt.figure(figsize=[4,2])
sns.countplot(data = tips, x = 'tip')
```

```python
plt.figure(figsize=[4,2])
sns.barplot(data = tips, x = 'sex' , y = 'tip')
```

```python
plt.figure(figsize=[4,2])
sns.countplot(data = tips, x = 'sex' , palette= 'magma')
```

```python
sns.boxplot(data = tips, x = 'day' , y = 'tip') 
```

```python
plt.figure(figsize=[4,2])
sns.histplot(tips['sex'])
```

```python
sns.boxplot(data = tips) 
```

```python
sns.boxplot(data = tips, x = 'day' , y = 'tip' , hue = 'sex') 
```

```python
sns.violinplot(data = tips, x = 'day' , y = 'tip',hue = 'sex', split='True' ) 
```

```python
sns.stripplot(data = tips, x = 'day' , y = 'tip' , palette = 'magma')
```

```python
sns.swarmplot(data = tips, x = 'day' , y = 'tip' , hue = 'sex')
```

```python
sns.boxplot(data = tips, x = 'day' , y = 'tip' , hue = 'sex' , palette= 'Spectral') 
sns.swarmplot(data = tips, x = 'day' , y = 'tip' , hue = 'sex')
```

```python
sns.lineplot(data = tips, x = 'total_bill' , y = 'tip')
```

```python
sns.scatterplot(data = tips, x = 'total_bill' , y = 'tip' ,hue = 'sex')
```

```python
sns.jointplot(data = tips, x = 'total_bill' , y = 'tip', kind='hex' )
```

```python
sns.pairplot(data = tips , hue = 'sex')
```

```python
g = sns.JointGrid(data = tips , x='total_bill',y='tip',)
g.plot( sns.scatterplot , sns.boxplot  )
```

```python
g = sns.JointGrid(data = tips )
g.map_diag (scatterplot)
g.map_upper(scatterplot )
g.map_lower(scatterplot)
```

```python
sns.heatmap(tips[['total_bill', 'tip' ,'size' ]].corr(),cmap='coolwarm')
```
