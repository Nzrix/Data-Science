```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dic = {'day': ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri','Sat'],
       'temp': [24,25,29,35,23,20,23]}
df = pd.DataFrame(dic)
x = df['day']
y = df['temp']
plt.plot(x, y,)
plt.show()
```

```python
plt.plot(x, y, 's:g', lw=3, mec='k')
plt.title('Temperature around this week')
plt.xlabel('Day')
plt.ylabel('Temp')
plt.xticks(rotation = 90)
plt.yticks([15,20,25,30,35,40])
plt.show()
```

```python
dic = {'day': ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri','Sat'],
       'temp': [24,25,29,35,23,20,23],
       'temp2':[10,14,9,12,11,15,17]}
df = pd.DataFrame(dic)
x = df['day']
y = df['temp']
z = df['temp2']
```

```python
df
```

```python
plt.plot(df['day'], df['temp'],'s:r',label = 'Temp1')
plt.plot(df['day'], df['temp2'],'o--b',label = 'Temp2')
plt.legend(loc = 'best')
```

```python
fig, axes = plt.subplots( ncols = 2, nrows = 2 )
axes[0,0].plot(x, y, 's:r')
axes[0,1].plot(x, z, 'o--b')
axes[1,0].plot(y, x, 'p-k')
axes[1,1].plot(z, x, '*-.g')
axes[0,0].set_title("week1")
axes[0,1].set_title("week2")
axes[1,0].set_title("week3")
axes[1,1].set_title("week4")
fig.tight_layout()
```

```python
fig= plt.figure()
axes1 = fig.add_axes([1,1,1,1])
axes1.plot(x, y, 's:r')

axes2 = fig.add_axes([0,0,0,1])
axes2.plot(x, z, 'o--b')

axes3 = fig.add_axes([0,0,1,1])
axes3.plot(y, x, 'p-k')

axes4 = fig.add_axes([1,1,1,1])
axes4.plot(z, x, '*-.g')
```
