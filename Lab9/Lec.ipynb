```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
df2 = df [['acceleration','weight']].dropna()

from sklearn.cluster import KMeans
model = KMeans(n_clusters=4,random_state=0)
model.fit(df2)


sns.scatterplot(data = df2 , x = 'acceleration', y = 'weight'
                ,hue=model.labels_,palette='Set1')
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1]
            ,color = 'k',marker='*')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df = sns.load_dataset("mpg")
df2 = df[['acceleration','weight']].dropna()

scaler.fit(df2)
df2[['acceleration','weight']] = scaler.transform(df2)

from sklearn.cluster import KMeans
model = KMeans(n_clusters=3,random_state=0)
model.fit(df2)

sns.scatterplot(data = df2, x = 'acceleration', y = 'weight', hue=model.labels_, palette='Set1')
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color= 'k', marker='*')
```

```python
df3 = df2
df3[['acceleration', 'weight']] = scaler.inverse_transform(df2)
center = scaler.inverse_transform(model.cluster_centers_)
plt.figure(figsize=[4,3])
sns.scatterplot(data = df3, x='acceleration', y='weight', hue=model.labels_, palette='Set1')

center = scaler.inverse_transform(model.cluster_centers_)
plt.scatter(center[:,0], center[:,1], color= 'k', marker='*')
```
