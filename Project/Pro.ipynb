```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dummy_data.csv')
```

```python
df.head()
```

```python
def life_stage(age):
    if age >0 and age <12:
        return "child"
    elif age>=12 and age<=20:
        return "teenage"
    elif age>20 and age<45:
        return "Adult"
    else:
        return "old"
```

```python
df["life_stage"] = df["age"].apply(life_stage)
```

```python
def Screen_time(number):
    if number<=2:
        return "Normal"
    elif number>2 and number<=4:
        return "Moderate"
    elif number>4 and number<=6:
        return "High"
    elif number>6:
        return "Extreme"
```

```python
df["Screen_time"] = df["time_spent"].apply(Screen_time)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Assuming df is your DataFrame containing the data
att = df[['age','time_spent']]
label = df['life_stage']

att_train, att_test, class_train, class_test = train_test_split(att, label, random_state=0, train_size=0.6)

scaler = StandardScaler()
scaler.fit(att_train)

att_train_scaled = scaler.transform(att_train)
att_test_scaled = scaler.transform(att_test)

model = KNeighborsClassifier(n_neighbors=3)
model.fit(att_train_scaled, class_train)

# Evaluate the model
train_score = model.score(att_train_scaled, class_train)
test_score = model.score(att_test_scaled, class_test)

print("Training Score:", train_score)
print("Test Score:", test_score)
```

```python
result = pd.concat([att_test,class_test],axis=1)
result['predict'] = model.predict(scaler.transform(att_test))
result
```
