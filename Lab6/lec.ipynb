```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,21)
y = np.random.randint(1,100,20)
```

```python
plt.scatter(x,y,marker = 'o', c='r' ,s=80)
```

```python
plt.figure(figsize = [6,4])
x = ['Mon','Tue','Wed','Thu','Fri']
y = [50,70,35,90,40]
plt.bar(x,y,color = 'b',alpha = 0.5,hatch = '\/-|',lw = 2,ec = 'k',ls = ':')
# plt.plot(x,y)
plt.title('Week')
```

```python
plt.figure(figsize = [6,4])
x = ['Mon','Tue','Wed','Thu','Fri']
y = [50,70,35,90,40]
plt.barh(x,y,color = 'r',alpha = 0.5,hatch = '\/-|',lw = 2,ec = 'k',ls = ':')
# plt.plot(x,y)
plt.title('Week')
```

```python
x = np.arange(1,6)
y1 = np.random.randint(1,50,5)
y2 = np.random.randint(1,50,5)
y3 = np.random.randint(1,50,5)
```

```python
plt.figure(figsize = [6,4])
plt.bar(x,y1,width=0.2)
plt.bar(x+0.2,y2,width=0.2)
plt.bar(x+0.4,y2,width=0.2)
```

```python
plt.figure(figsize = [4,2 ])
bw = 0.2
plt.bar(x,y1,width=bw,label='A')
plt.bar(x+bw,y2,width=bw,label='B')
plt.bar(x+bw*2,y3,width=bw,label='C')
plt.xticks(x+bw,['2000','2001','2002','2003','2004'])
plt.legend(loc = 'best')
plt.show()
```

```python
dic = {"Match" : [10,40,78,55,60],
       "Eng" : [99,23,56,87,65],
       "Thai": [78,50,60,34,25]
       }
df = pd.DataFrame(dic)
data = [df['Match'],df['Eng'],df['Thai']]
plt.boxplot(data,vert=True,widths=0.3,labels=["Match", "Eng", "Thai"],patch_artist=True)
plt.show()
```
